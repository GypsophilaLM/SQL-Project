```python
# URLs for the datasets
census_data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01"

schools_data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01"

crime_data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01"
```

```python
# Load data into Pandas DataFrames
census_df = pd.read_csv(census_data_url)
schools_df = pd.read_csv(schools_data_url)
crime_df = pd.read_csv(crime_data_url)
```

```python
# Establish connection to SQLite database
connection = sqlite3.connect('FinalDB.db')
```

```python
# Drop existing table
connection.execute('DROP TABLE IF EXISTS CENSUS_DATA')
connection.execute('DROP TABLE IF EXISTS CHICAGO_PUBLIC_SCHOOLS')
connection.execute('DROP TABLE IF EXISTS CRIME_DATA')
```

```python
# Use Pandas to write DataFrames to SQLite tables
# 每次将数据存储到SQLite数据库中都需要运行一次
census_df.to_sql('CENSUS_DATA', connection, index=False, if_exists='replace')
schools_df.to_sql('CHICAGO_PUBLIC_SCHOOLS', connection, index=False, if_exists='replace')
crime_df.to_sql('CRIME_DATA', connection, index=False, if_exists='replace')
```

```python
# Close the connection
connection.close()
```

```python
# Find the total number of crimes recorded in the CRIME table
%sql SELECT COUNT(*) AS total_crimes \
FROM crime_data
```

```python
# List community area names and numbers with per capita income less than 11000
%sql SELECT community_area_name, community_area_number \
    FROM census_data \
    WHERE per_capita_income < 11000;
```

```python
# List all case numbers for crimes involving a child
%sql SELECT case_number \
    FROM crime_data \
    WHERE primary_type LIKE '%CHILD%' OR description LIKE '%CHILD%'
```

```python
# List all kidnapping crimes involving a child
%sql SELECT case_number \
    FROM crime_data \
    WHERE primary_type = 'KIDNAPPING'
```

```python
# List the kind of crimes that were recorded at schools
%sql SELECT case_number, location_description \
    FROM crime_data \
    WHERE location_description LIKE '%SCHOOL%'
```

```python
# show the average safety score for Middle School
%sql SELECT AVG(safety_score) AS avg_safety_score \
    FROM chicago_public_schools \
    WHERE "Elementary, Middle, or High School" LIKE '%MS%';
```

```python
# List 5 community areas with highest % of households below poverty line
%sql SELECT community_area_name, \
            AVG(percent_households_below_poverty) AS avg_percent_below_poverty \
    FROM census_data \
    GROUP BY community_area_name \
    ORDER BY avg_percent_below_poverty DESC \
    LIMIT 5;
```

```python
# Which community area is most crime prone? Display the coumminty area number only
%sql SELECT community_area_number \
     FROM crime_data \
     GROUP BY community_area_number \
     ORDER BY COUNT(*) DESC \
     LIMIT 1;
```

```python
# Use a sub-query to find the name of the community area with highest hardship index
%sql SELECT community_area_name \
     FROM census_data \
     WHERE hardship_index = (SELECT MAX(hardship_index) FROM census_data);
```

```python
# Use a sub-query to determine the Community Area Name with most number of crimes
%sql SELECT community_area_name \
     FROM census_data \
     WHERE community_area_number = (SELECT community_area_number \
                                    FROM crime_data \
                                    GROUP BY community_area_number \
                                    ORDER BY COUNT(*) DESC \
                                    LIMIT 1);
```
